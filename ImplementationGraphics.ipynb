```python
# Retransmisiones prueba 1

import json
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos JSON generados por iperf3
with open('resultado1.json', 'r') as f:
    data = json.load(f)

# Extraer los intervalos
intervals = data['intervals']

# Crear listas para almacenar los datos
times = []
bps = []  # Bits por segundo
retransmits = []  # Retransmisiones TCP

# Extraer los datos de cada intervalo
for interval in intervals:
    times.append(interval['sum']['end'])
    bps.append(interval['sum']['bits_per_second'])
    retransmits.append(interval['sum']['retransmits'])

# Convertir a DataFrame para análisis más sencillo
df = pd.DataFrame({
    'Tiempo (s)': times,
    'Tasa de bits (bps)': bps,
    'Retransmisiones': retransmits
})

# Graficar las retransmisiones TCP
plt.figure(figsize=(10, 6))
plt.plot(df['Tiempo (s)'], df['Retransmisiones'], label='Retransmisiones TCP', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Número de retransmisiones')
plt.title('Comportamiento TCP: Retransmisiones vs. Tiempo')
plt.grid(True)
plt.legend()
plt.show()

############################################################
# Cambiar limites de los ejes para mejor visualización

# Graficar las retransmisiones TCP
plt.figure(figsize=(10, 6))
plt.plot(df['Tiempo (s)'], df['Retransmisiones'], label='Retransmisiones TCP', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Número de retransmisiones')
plt.title('Comportamiento TCP: Retransmisiones vs. Tiempo')
# Limitar el eje X y Y
plt.xlim(500, 600)
plt.ylim(0, 10)

plt.grid(True)
plt.legend()
plt.show()


```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Función para cargar y procesar un archivo JSON
def procesar_json(filename):
    with open(filename, 'r') as f:
        data = json.load(f)

    # Extraer intervalos de tiempo y métricas
    intervals = data['intervals']
    times = []
    bps = []  # Bits por segundo
    retransmits = []  # Retransmisiones TCP

    for interval in intervals:
        times.append(interval['sum']['end'])
        bps.append(interval['sum']['bits_per_second'])
        retransmits.append(interval['sum']['retransmits'])

    # Devolver un DataFrame con los datos procesados
    return pd.DataFrame({
        'Tiempo (s)': times,
        'Tasa de bits (bps)': bps,
        'Retransmisiones': retransmits
    })

# Cargar los tres archivos JSON
df1 = procesar_json('resultado1.json')
df2 = procesar_json('resultado2.json')
df3 = procesar_json('resultado3.json')

# Comparar las retransmisiones TCP de las tres pruebas
plt.figure(figsize=(10, 6))
plt.plot(df1['Tiempo (s)'], df1['Retransmisiones'], label='Prueba 1', color='blue')
plt.plot(df2['Tiempo (s)'], df2['Retransmisiones'], label='Prueba 2', color='green')
plt.plot(df3['Tiempo (s)'], df3['Retransmisiones'], label='Prueba 3', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Retransmisiones')
plt.title('Comparación de Retransmisiones TCP entre las 3 Pruebas')
plt.grid(True)
plt.legend()
plt.show()

############################################################
# Cambiar limites de los ejes para mejor visualización

# Comparar las retransmisiones TCP de las tres pruebas
plt.figure(figsize=(10, 6))
plt.plot(df1['Tiempo (s)'], df1['Retransmisiones'], label='Prueba 1', color='blue')
plt.plot(df2['Tiempo (s)'], df2['Retransmisiones'], label='Prueba 2', color='green')
plt.plot(df3['Tiempo (s)'], df3['Retransmisiones'], label='Prueba 3', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Retransmisiones')
plt.title('Comparación de Retransmisiones TCP entre las 3 Pruebas')
# Limitar el eje X y Y
plt.xlim(40, 80)
plt.ylim(0, 6)
plt.grid(True)
plt.legend()
plt.show()
```
